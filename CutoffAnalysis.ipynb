```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

```python
x=pd.read_html("https://collegedunia.com/exams/tnea/cutoff")
y=x[1]
y
```

```python
courses=y["B.Tech Programs"]
npclose=np.array(y["Closing Cutoff Marks"])
npopen=np.array(y["Opening Cutoff Marks"])
range=npopen-npclose
```

```python
plt.xlabel('Cutoff Range')
plt.ylabel('B.Tech Programs')
plt.title('B.Tech Programs Cutoff Ranges')
plt.barh(courses,range,left=npclose,height=0.4,color="green")
plt.show()
```
